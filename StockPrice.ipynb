```python
import pandas as pd
import numpy as np
```

```python
#Storing the names of the columns that are to be converted into numeric form in a list
columns_to_convert = ['Change', '365 Day Percentage Change', '30 Day Percentage Change']
```

```python
#Converting the columns into numeric and checking if there are any '-'  values and converting them to NaN
df[columns_to_convert] = df[columns_to_convert].replace('-', pd.NA)
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Fill missing values in 'Change' by calculating 'Last Traded Price' - 'Previous Close'
df['Change'] = df['Change'].fillna(df['Last Traded Price'] - df['Previous Close'])
```

```python
# Fill missing values in '365 Day Percentage Change' with the column's mean
df['365 Day Percentage Change'] = df['365 Day Percentage Change'].fillna(df['365 Day Percentage Change'].mean())
```

```python
# Fill missing values in '30 Day Percentage Change' with the column's mean
df['30 Day Percentage Change'] = df['30 Day Percentage Change'].fillna(df['30 Day Percentage Change'].mean())
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Dropping unncessary column
X = df.drop(['Company Name', 'Symbol', 'Industry', 'Series', 'Last Traded Price', 'Percentage Change'], axis=1)
```

```python
X.dtypes
```

```python
#Storing the target ['Last Traded Price']
y_ltp = df['Last Traded Price']
```

```python
y_ltp
```

```python
#We are performing MinMaxScaler here
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
X_sc = scaling.fit_transform(X)
```

```python
X_sc
```

```python
y_ltp
```

```python
#Splitting the model into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sc, y_ltp, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Initialize and train the Linear Regression model
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

# Make predictions and evaluate
y_pred = linear_regressor.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Linear Regression - Mean Absolute Error (MAE): {mae}")
print(f"Linear Regression - Root Mean Squared Error (RMSE): {rmse}")
```

```python
#Getting the prediction score
linear_regressor.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a DataFrame for plotting
results_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

# Plot Predicted vs. Actual values
plt.figure(figsize=(10, 6))
sns.scatterplot(data=results_df, x='Actual', y='Predicted', alpha=0.6)
plt.plot([results_df['Actual'].min(), results_df['Actual'].max()], 
         [results_df['Actual'].min(), results_df['Actual'].max()], 
         color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Values')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Initialize and train the Random Forest Regressor
random_forest_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
random_forest_regressor.fit(X_train, y_train)

# Make predictions
y_pred_rf = random_forest_regressor.predict(X_test)

# Evaluate the Random Forest model
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))

print(f"Random Forest - Mean Absolute Error (MAE): {mae_rf}")
print(f"Random Forest - Root Mean Squared Error (RMSE): {rmse_rf}")
```

```python
#Getting Random Forest score
random_forest_regressor.score(X_test, y_test)
```

```python
#Handling categorical data and filling the missing values

columns_to_convert = ['Change', '365 Day Percentage Change', '30 Day Percentage Change']
df[columns_to_convert] = df[columns_to_convert].replace('-', pd.NA)
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Fill missing values without inplace
df['Change'] = df['Change'].fillna(df['Last Traded Price'] - df['Previous Close'])
df['365 Day Percentage Change'] = df['365 Day Percentage Change'].fillna(df['365 Day Percentage Change'].mean())
df['30 Day Percentage Change'] = df['30 Day Percentage Change'].fillna(df['30 Day Percentage Change'].mean())

# Convert '365 Day Percentage Change' and '30 Day Percentage Change' to binary (1 for positive, 0 for non-positive)
df['365 Day Percentage Change'] = np.where(df['365 Day Percentage Change'] > 0, 1, 0)
df['30 Day Percentage Change'] = np.where(df['30 Day Percentage Change'] > 0, 1, 0)
```

```python
#Handling categorical data and filling the missing values

df['Percentage Change'] = df['Percentage Change'].replace('-', pd.NA)
df['Percentage Change'] = df['Percentage Change'].apply(pd.to_numeric, errors='coerce')
df['Percentage Change'] = df['Percentage Change'].fillna((df['Change'] / df['Previous Close']) * 100)
df['Percentage Change'] = np.where(df['Percentage Change'] > 0, 1, 0)
```

```python
df.isnull().sum()
```

```python
X_percent = df.drop(columns=['Company Name', 'Symbol', 'Industry', 'Series', 'Percentage Change'])
```

```python
X_percent.dtypes
```

```python
from sklearn.preprocessing import RobustScaler

# Initialize the RobustScaler
scaling = RobustScaler()

# List of columns to scale
scaling_columns = ['Open', 'High', 'Low', 'Previous Close', 'Last Traded Price', 
                   'Change', 'Share Volume', 'Value (Indian Rupee)', 
                   '52 Week High', '52 Week Low']

# Create a copy of X to store the scaled values
X_sc_percent = X_percent.copy()

# Apply Robust Scaling to the selected columns
X_sc_percent[scaling_columns] = scaling.fit_transform(X_percent[scaling_columns])

# Display the scaled DataFrame
X_sc_percent
```

```python
y_percent = df['Percentage Change']
y_percent
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC  # Import SVCv
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Train SVM model
svm_model = SVC()
#svm_model.fit(X_train, y_train)
lsvm_model = LogisticRegression()
#lsvm_model.fit(X_train, y_train)

# Train Random Forest model
rf_model = RandomForestClassifier()
# Define the number of folds for cross-validation
num_folds = 20

# Cross-validate SVM model
svm_scores = cross_val_score(svm_model, X_sc_percent, y_percent, cv=num_folds, scoring='accuracy')
print("SVM Cross-Validation Accuracy Scores:\n", svm_scores)
print("SVM Mean Cross-Validation Accuracy:", svm_scores.mean())

# Cross-validate LSVM (Logistic Regression) model
lsvm_scores = cross_val_score(lsvm_model, X_sc_percent, y_percent, cv=num_folds, scoring='accuracy')
print("\nLSVM Cross-Validation Accuracy Scores:\n", lsvm_scores)
print("LSVM Mean Cross-Validation Accuracy:", lsvm_scores.mean())

# Cross-validate Random Forest model
rf_scores = cross_val_score(rf_model, X_sc_percent, y_percent, cv=num_folds, scoring='accuracy')
print("\nRandom Forest Cross-Validation Accuracy Scores:\n", rf_scores)
print("Random Forest Mean Cross-Validation Accuracy:", rf_scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score

# Train SVM model
svm_model = SVC()
# svm_model.fit(X_train, y_train)
lsvm_model = LogisticRegression()
# lsvm_model.fit(X_train, y_train)

# Train Random Forest model
rf_model = RandomForestClassifier()

# Define the number of folds for cross-validation
num_folds = 20

# Cross-validate SVM model
svm_recall_scores = cross_val_score(svm_model, X_sc_percent, y_percent, cv=num_folds, scoring='recall')
print("SVM Cross-Validation Recall Scores:\n", svm_recall_scores)
print("SVM Mean Cross-Validation Recall:", svm_recall_scores.mean())

# Cross-validate LSVM (Logistic Regression) model
lsvm_recall_scores = cross_val_score(lsvm_model, X_sc_percent, y_percent, cv=num_folds, scoring='recall')
print("\nLSVM Cross-Validation Recall Scores:\n", lsvm_recall_scores)
print("LSVM Mean Cross-Validation Recall:", lsvm_recall_scores.mean())

# Cross-validate Random Forest model
rf_recall_scores = cross_val_score(rf_model, X_sc_percent, y_percent, cv=num_folds, scoring='recall')
print("\nRandom Forest Cross-Validation Recall Scores:\n", rf_recall_scores)
print("Random Forest Mean Cross-Validation Recall:", rf_recall_scores.mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming svm_scores, lsvm_scores, and rf_scores are your cross-validation scores

# Prepare data for plotting
data = {
    'Model': ['SVM'] * len(svm_scores) + ['Logistic Regression'] * len(lsvm_scores) + ['Random Forest'] * len(rf_scores),
    'Accuracy': np.concatenate([svm_scores, lsvm_scores, rf_scores])
}

# Create a DataFrame from the data
import pandas as pd
df = pd.DataFrame(data)

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Box Plot
plt.subplot(1, 3, 1)
sns.boxplot(x='Model', y='Accuracy', data=df)
plt.title('Box Plot of Model Accuracy')

# Bar Plot
plt.subplot(1, 3, 2)
mean_accuracies = df.groupby('Model').mean().reset_index()
sns.barplot(x='Model', y='Accuracy', data=mean_accuracies)
plt.title('Mean Accuracy of Models')

# Show plots
plt.tight_layout()
plt.show()
```
